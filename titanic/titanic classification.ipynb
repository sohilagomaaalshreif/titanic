```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
```

```python
features = df.drop(columns=["Cabin" , "Name" , "Embarked" , "Ticket","Survived","Sex"])
goal=df["Survived"]
features
goal
```

```python
# import the modules we need
from sklearn.preprocessing import Imputer
# make a new imputer 
imputer = Imputer(missing_values='NaN', strategy='mean',axis=0)
features[['Age']] = imputer.fit_transform(features[['Age']])
features
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set, goal_train, goal_test = train_test_split(features, goal, test_size =1/3, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(train_set, goal_train)
Y_pred = random_forest.predict(test_set)
random_forest.score(train_set, goal_train)
acc_random_forest = round(random_forest.score(train_set, goal_train) * 100, 2)
acc_random_forest
```
